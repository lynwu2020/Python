```python
## axis的理解：axis意味着最外层门是几个；如[[1,2],[3,4]]的axis为2
##axis=0的计算，表示对最外层[ ]里的最大单位块做块与块之间的运算,同时移除最外层[ ]：
#二维情形：axis=0代表在跨行（down),沿着每一列或行标签\索引值向下按列计算；axis=1代表跨列（across)，按行计算（想象成一个表格）
import numpy as np
a= np.array([1,2,3])    
print(a.sum(axis = 0))
b= np.array([[1,2],[3,4]])  ##最外层里的最大单位块为[1,2]和[3,4],运算sum得[[4,6]],移除最外层为[4,6]
print(b.sum(axis = 0))
## axis= 1表示对第二外层[]里的最大单位块做块与块之间的运算,同时移除第二外层[]：
print(b.sum(axis = 1))   #第二外层内为 1，2和3，4，分别运算得[[3,7]],移除最外层得[3,7]
## stack(arrays,axis=0)
c=np.array([4,5,6])    
print('*'*60) 
d=np.stack((a,c),axis=0)  # 合并数组为[[1, 2, 3], [4, 5, 6]]，最外层里的最大单位块为[1,2,3]和[3,4,5],直接堆叠；
e=np.stack((a,c),axis=1)  # 合并数组为[[1, 2, 3], [4, 5, 6]]，第二层里的最大单位块为1,2,3和3,4,5,直接堆叠；
print(d,'\n') 
print(e,'\n') 
```

```python
a=[[1],[2],[3]]
b=[[1],[2],[3]]
c=[[1],[2],[3]]
f=np.hstack((a,b,c))       #水平(按列顺序)把数组给堆叠起来
g=np.vstack((a,b,c))       #垂直(按行顺序)把数组给堆叠起来
print(f,'\n')  
print(g,'\n') 
```
