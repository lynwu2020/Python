## 编码
默认情况下，Python 3 源码文件以 UTF-8 编码，所有字符串都是 unicode 字符串。 当然你也可以为源码文件指定不同的编码。

## 标识符
- 第一个字符必须是字母表中字母或下划线 _ ，Python 3 中，可以用中文作为变量名，非 ASCII 标识符也是允许的了。
- 标识符的其他的部分由字母、数字和下划线组成。
- 标识符对大小写敏感。

## 行与缩进
- python最具特色的就是使用缩进来表示代码块，不需要使用大括号 {} 。
- 缩进的空格数是可变的，但是同一个代码块的语句必须包含相同的缩进空格数。建议始终4个空格的缩进!
- 函数之间或类的方法之间用空行分隔，表示一段新的代码的开始。类和函数入口之间也用一行空行分隔，以突出函数入口的开始。
- print输出默认换行，如果要实现不换行需要在变量末尾加上 end=""；
- Python允许用'''...'''的格式表示多行内容；
- Python 通常是一行写完一条语句，但如果语句很长，我们可以使用反斜杠(\)或者括号来实现多行语句；同一行多行语句用;分割；

## 变量和格式化：
- 方程式=左边即变量，给变量名赋值即可创建该变量并完成赋值，变量可以是任意数据类型。
- 变量名必须是大小写英文、数字和_的组合，且不能用数字开头；不能使用保留字；
- 多个变量赋值：如a = b = c = 1；a, b, c = 1, 2, "runoob"；
- 格式化方式和C语言是一致：%d,%f,%s,%x(十六进制),如 print('%s,you\'ve improved %.2f%% than last year'%(name,r))

## 标准数据类型
Python3 中有六个标准的数据类型：
不可变数据（3 个）：Number（数字）、String（字符串）、Tuple（元组）；
可变数据（3 个）：List（列表）、Dictionary（字典）、Set（集合）
特殊字符：空值 None
数据类型转换函数：int(),float(),str()
- Number（数字）
    - 四种类型：int (整数，长整型)、bool (布尔，True、False两种值)、float (浮点数)和complex (复数，如1 + 2j)。
- String（字符串）
    - 单引号'或双引号"或者三引号括起来的任意文本（python中单引号和双引号使用基本相同）
    - 字符串输出：字符串内部包含',可用""括起来;字符串内部包含"，可用‘’‘ ’‘’括起来，用转义字符\转义;
    - 转义字符\可以转义很多字符，比如\n表示换行，\t表示制表符；字符\本身也要转义，\\表示的字符\;使用r可以让\不发生转义，如r"\nb";
    - 加号 + 是字符串的连接符,星号 * 2表示复制当前字符串2次；
    - 字符串可视为列表，索引切片规则与列表相同，切片为左闭右开区间
- list（列表）
    - 用[]标识，有序对象，支持字符、数字、字符串、可嵌套列表，如a=[0,1,2,[1,2],3]
    - 每个元素分配顺序 0-N 和倒叙 N--1 两种索引来标识元素位置；索引切片方式如a[0],a[0:5],a[-4]
    - 生成器range()，range(0,10,2)表示取值范围为0=<i<10,且步长为2的整数；配合list()生成列表，如list(range(1,4,2));
    - 常用操作：添加
- Tuple（元组）
    - 用()标识，有序对象，不可二次赋值（只读）如(1,2,3,b)，字符串属于不可变列表；
    - 元组中只包含一个元素时需要在元素后面添加逗号，如(50,)
    - 元组不能单独删除内部元素，可以用 del 删除整个元组，如 del tup；
- Set（集合）
    - 是一个无序的不重复元素序列，可以使用大括号{ }或者 set() 函数创建集合；
- Dictionary（字典）
    - 用{}标识，无序对象，由索引key和value值的键值对组成；如{'a':1,'b'=2}
    - 同一个Key只能出现一次；key 必须是字符串，Value 可以是任意对象

### 数字字符串常用操作

```python
s=str='Runoob'
type(s)
# 字符串索引切片输出操作
print(str[0:-1]) # 左闭右开区间，输出索引 0到-2所有值；
print(s[2:]) # 输出从索引2及之后开的所有值
print(s[:2]) # 输出从索引2之前的所有值；
print(s[:6:2]) # 输出从索引6之前步长的所有值； 右边开区间即024
print(s[1::2]) # 输出从索引6之前步长的所有值； 左边闭区间即135
print(len(s)) #求列表元素个数/长度，其他常见函数如max,min,sum
print(s * 2) # 输出字符串两次 
print(s + '你好') # 连接字符串 
print('*'*60)
#转义输出特殊字符
print('hello\'runoob') # 使用反斜杠(\)转义特殊字符 
print(r'hello\nrunoob') # 在字符串前面添加一个r，表示原始字符串，不会发生转义
print('''hello ''
nrunoob''') # ’‘’输出多行内容，字符串内部包含"或者‘，可用‘’‘ ’‘’括起来输出；
print('*'*60)
#替换
s1=s.replace('o','n',1)  #替换1个o
print(s1)
print('*'*60)
#拆分字符串
s2=s1.split('o')
print(s2)
print('*'*60)
#连接
m='_'
s3=m.join(s2)
print(s3)
print('*'*60)
#大小写
print(s.upper())
print(s.lower())
print(s.swapcase())#大小写互换
print(s.capitalize())#首字母大写
print('*'*60)
#其他
print(s.startswith('b'),s.endswith('b')) #判断开头结尾字母
print(s.isnumeric())#是否全数字
print(s.isalpha()) #是否全字母
print(s.isdigit()) #是否数字
print(s.isspace()) #是否数字
print(s.rstrip()) #删除字符末尾空格
print('*'*60)
#格式化
s4='%s is 好人'%s
print(s4)
s4='%s和%s is 好人'%(s,s1)
print(s4)
m=3.141592657
print('PI is %.4f'%m) #保留 4 位小数四舍五入
print('PI is %+i'%m) #取整数部分,显示+号
print('PI is % +i'%m) #空格和正号只显示一个
print('PI is %.2e'%m) #科学计数法
print('PI is %.2g'%m) #小数位数少自动识别用浮点，复杂时自动用科学计数法,数字为取有效数字而非小数位数
print('*'*60)
#更强大格式化
print('{},今天天气怎么样?{}'.format('你好','非常好'),'\n') #任意替换
print('{2}{1}{0}'.format('a','b','c'),'\n') #可以指定元素顺序
print('{},今天天气怎么样?{情况}'.format('你好',情况='好！'),'\n') #可以使用变量指示
print('{:.2f}'.format(4.12312))
```

## 序列、元组通用操作

```python
#列表和生成器
team1=['leo','araon']
classa = ['Michael', 'Bob',team1, 'Tracy']
list4=list(range(0,6,2)) 
#替换元素
classa[2][1] = 'Sarah' #元素赋值
print(list4,classa,classa[2],type(classa))
print('----------------------------------------\n')
#添加元素
list3=team1*2+[0,1,2,3,4]  #序列的链接，列表+列表，序列的重复*N
classa.append(['Leon','Nick'])   #在列表最后添加一个元素，支持所有数据类型，输入新列表会整个作为一个元素添加；
classa.extend(['Leon','Nick']) #在列表最后逐个添加元素，输入新列表则会按照列表元素逐个添加；
classa.insert(0,'Adam')  #insert在指定索引插入值,右侧值索引顺延1；支持所有数据类型；
print(list3,classa,classa[2],type(classa))
print('----------------------------------------\n')
#删除元素
classa.pop(0)  #删除指定索引位置元素
list3.remove(4) #删除值为 1 的元素
del classa[0:1] #删除0-1切片内元素
list4=list(range(0,6,2));list4.clear() #清空列表
print(list4,list3,classa,classa[2],type(classa))
print('----------------------------------------\n')
#复制列表
a=list(range(0,10,2))
b=a
print(b)
a[2]=100
print(a)
print(b) ##由于m a 共用一个索引，所以a[2]和 m[2]指向同一个对象，故更改任何一个，两个均变化；
print('----------------------------------------\n')#下面是正确姿势
b=b.copy()
b[2]=10
print(a)
print(b)
print('----------------------------------------\n')
#列表的排序 .sort()方法
a.sort()
b.sort(reverse=True)
print(a)
print(b)
c=sorted(b) # sorted()函数排序
print(c)
print('----------------------------------------\n')
#序列索引切片
1 not in a # 判断是否属于序列，列表索引切片规则跟 str 相同，函数max,min,sum适用
print(a.index(2)) #返回该值第一个索引
print(a.count(2)) #数该值出现次数
print(a[0:-1]) # 左闭右开区间，输出索引 0到-2所有值；
print(a[:6:2]) # 输出从索引6之前步长的所有值； 右边开区间即024
print(len(a)) #求列表元素个数/长度，其他常见函数如max,min,sum
print(a * 2) # 输出两次 
print(a + ['你好']) # 连接两个序列
```

### 字典通用操作

```python
#字典申明
dic1={'a':18,'b':19,'c':20}  #直接创建字典
dic2=dict(jack=20,marry=19,jone=25) #dict函数创建字典
list1=[['e',18],['f',19],['g',15],['h',17]]; dic3=dict(list1)#通过列表+dict函数创建字典
list2=['i','j','k','l','m','n']; dic4=dict.fromkeys(list2,0) #通过列表+dict.fromkeys()创建keys,默认 value 为 none
#字典变更
dic1['d']='任意添加' #任意添加key-value
dic2['c']='任意修改' #任意修改key-value
del dic1['a'] #任意删除
print(dic1)
print(dic2)
print('----------------------------------------\n')
#复制/更新/合并字典
dic1.update(dic2)  #将 dic2 添加到dic1
dic=dic1.copy() #跟列表相似，为避免两个列表同步修改，复制列表需要更新一个新的指向；
dic.update({'d':22})
print(dic)
print('----------------------------------------\n')
#元素访问:
print(len(dic1)) #len()函数计算长度
print('b' in dic1) #只能根据key判断是否包含 
print(dic3.get('e',print('nothing'))) #.get()方法获得该 key的 value，无法找到则默认返回none，可自定义 print
print(list(dic.keys())) #.keys获得所有的 key，type为 range，可以用 list()形成列表
print(list(dic.values()))#.values获得所有的 value，type为 range，可以用 list()形成列表
print(list(dic.items())) #.items得所有的 元素，type为 range，可以用 list()形成列表
for keys in dic:  #输出keys
    print(keys)
for values in dic.values():  #通过.values输出values
    print(values)
for keys2 in dic:            #通过key+索引方式输出 values
    print(dic[keys2])
for items in dic.items():   #直接输出 items
    print(items)
for (k,v) in dic.items():   #通过(k,v)遍历 items 可分别获得k,v
    print('keys为%s,values为:%s'%(k,v))
```
